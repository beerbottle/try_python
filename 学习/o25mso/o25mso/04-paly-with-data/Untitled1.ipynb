```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0,5.0)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.plot(x,y1,'b--',label = 'sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x,y2,'r--',label = 'sin(pi*2x)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend()
```
