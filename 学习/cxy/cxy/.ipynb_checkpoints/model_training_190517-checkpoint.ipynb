```python
'''（2）计算特征的psi,删除psi较高的特征'''
def psi_categ(actual,expected):
    acnt = actual.value_counts()
    ecnt = expected.value_counts()
    arate=actual.value_counts()/len(actual)
    erate=expected.value_counts()/len(expected)

    psi = np.sum((arate - erate)*np.log(arate/erate))
    
    return psi

def psi_contins(actual,expected,quant=10):
    minv = min(min(actual),min(expected))
    maxv = max(max(actual),max(expected))
    step = 1.0*(maxv-minv)/quant
    acnt = []
    ecnt = []
    s,e = minv,minv+step
    act = np.array(actual)
    ex = np.array(expected)

    while e <= maxv and step!=0:
        acnt.append(((act>=s) & (act < e)).sum())
        ecnt.append(((ex>=s )& (ex < e)).sum())
        s = e
        e = e+step

    arate = np.array(acnt)/ len(actual)
    erate = np.array(ecnt)/ len(expected)

    arate[arate==0] = 0.000001
    erate[erate==0] = 0.000001

    psi = np.sum((arate - erate)*np.log(arate/erate))

    return psi
```

```python
data_X = data.drop(['Unnamed: 0','label'],axis=1)
data_corr = data_X.corr()

#改变相关性矩阵形式
for i in range(len(data_corr)):
    for j in range(len(data_corr)):
        if i<=j:
            data_corr.iloc[i,j]=np.nan 
            
data_corr_col = data_corr.columns.tolist()

#找出相互之间相关性较高的列
data_high_corr=pd.Series()
for i in data_corr_col:
    data_high_corr = pd.concat([data_high_corr,pd.Series(data_corr[i][data_corr[i]>0.7].index)],axis=1)
data_corr_col = data_corr.columns.tolist()
data_high_corr = data_high_corr.iloc[:,1:]
data_high_corr.columns=data_corr_col
data_high_corr = data_high_corr.dropna(axis=1, how='all') #删除表中全部为NaN的列
data_high_corr_col = data_high_corr.columns.tolist()

#计算特征IV值
high_corr_categ_col = list(set(data.columns).intersection(categ_col))
high_corr_contins_col = list(set(data.columns).intersection(contins_col))

#等分10份
def bin_equal(contns_feature,bin_num):
    group_names =range(bin_num)
    bins=np.linspace(min(contns_feature)-0.0000001,max(contns_feature),bin_num+1)
    contns_feature = pd.cut(contns_feature,10,labels=group_names)
    return contns_feature

for feat in high_corr_contins_col:
    data[feat] = bin_equal(data[feat],10)
    
    
def info_values(data_,DepVarName,features):
    def cal_iv(x):
        df=data_.groupby(x,as_index=False).agg({DepVarName:[np.sum,np.size]})
        df.columns=["index","bad","total"]
        df.set_index(["index"], inplace=True)
        df['good'] = df['total'] - df['bad']
        woe = np.log( (df['bad']/df['bad'].sum()+0.0000001)/(df['good']/df['good'].sum()+0.0000001) )
        diff = (df['bad']/df['bad'].sum()) - (df['good']/df['good'].sum())
        return woe,(diff*woe).sum()

    woe = [[feat,cal_iv(feat)[0]] for feat in features]
    iv = pd.DataFrame([[feat,cal_iv(feat)[1]] for feat in features],columns=["Feature",'IV'])
    #return iv[iv['IV']>iv_alpha].sort_values(by='IV',ascending=False),woe
    return iv

high_corr_categ_var_iv = info_values(data,'label',high_corr_categ_col)
high_corr_contins_var_iv = info_values(data,'label',high_corr_contins_col)
high_corr_var_iv = pd.concat([high_corr_categ_var_iv,high_corr_contins_var_iv],axis=0)

remove_high_corr_var = []
for i in data_high_corr.columns.tolist():
    col_iv = high_corr_var_iv.IV[high_corr_var_iv.Feature==i]
    col_iv = np.float(col_iv)
    for j in range(len(data_high_corr)):
        if type(data_high_corr.loc[j,i]).__name__=='str':
            row_iv = high_corr_var_iv.IV[high_corr_var_iv.Feature==data_high_corr.loc[j,i]]
            row_iv = np.float(row_iv)
            if col_iv<row_iv:
                remove_high_corr_var.append(i)
                break
            else:
                remove_high_corr_var.append(data_high_corr.loc[j,i])
```

```python
"""参数优化"""
def modelfit(alg, dtrain, predictors,useTrainCV=True,cv_folds=5, early_stopping_rounds=50):
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(dtrain[predictors].values, label=dtrain["label"].values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
                        metrics='auc', early_stopping_rounds=early_stopping_rounds)
    alg.set_params(n_estimators=cvresult.shape[0])
    alg.fit(dtrain[predictors], dtrain["label"],eval_metric='auc')

    dtrain_predictions = alg.predict(dtrain[predictors])
    dtrain_predprob = alg.predict_proba(dtrain[predictors])[:,1]

    print ("\nModel Report")
    print ("Accuracy : %.4g" % metrics.accuracy_score(dtrain['label'].values, dtrain_predictions))
    print ("AUC Score (Train): %f" % metrics.roc_auc_score(dtrain['label'], dtrain_predprob))

    feat_imp = pd.Series(alg.feature_importances_).sort_values(ascending=False)
    feat_imp.plot(kind='bar', title='Feature Importances')
    plt.ylabel('Feature Importance Score')
    """
    plot_importance(xgb1)
    pyplot.show()
    """
```

```python
def grid_search_result(gsearch):
    means_train_score = gsearch.cv_results_['mean_train_score']
    means_test_score = gsearch.cv_results_['mean_test_score']
    std_train_score = gsearch.cv_results_['std_train_score']
    std_test_score = gsearch.cv_results_['std_test_score']
    params = gsearch.cv_results_['params']
    cv_result = pd.DataFrame({'params':params,'means_train_score':means_train_score,'means_test_score':means_test_score,'std_train_score':std_train_score,
                 'std_test_score':std_test_score,'diff':abs(means_train_score-means_test_score)/means_train_score*100})
    cv_result = cv_result.sort_values(['diff','means_test_score'],ascending=[True,False])
    return cv_result
```

```python
def plot_psi(actual,expected,title = 'PSI',quant=10):
    minv = min(min(actual),min(expected))
    maxv = max(max(actual),max(expected))
    step = 1.0*(maxv-minv)/quant
    acnt = []
    ecnt = []
    s,e = minv,minv+step
    act = np.array(actual)
    ex = np.array(expected)

    while e <= maxv:
        acnt.append(((act>=s) & (act < e)).sum())
        ecnt.append(((ex>=s )& (ex < e)).sum())
        s = e
        e = e+step

    arate = np.array(acnt)/ len(actual)
    erate = np.array(ecnt)/ len(expected)

    arate[arate==0] = 0.000001
    erate[erate==0] = 0.000001

    psi = np.sum((arate - erate)*np.log(arate/erate))
    #psi = psi.sum()

    x1 = np.linspace(0,len(acnt)-1,len(acnt)) - 0.2
    x2 = np.linspace(0,len(acnt)-1,len(acnt)) + 0.2
    plt.bar(x1, arate*100, alpha=1, width = 0.4, facecolor = 'orange', edgecolor = 'white', label='train', lw=0.2)
    plt.bar(x2, erate*100, alpha=0.8, width = 0.4, facecolor = 'blue', edgecolor = 'white', label='test', lw=0.2)
    plt.legend(fontsize=15)
    
    
    lx=[]
    for i in range(11):
        lx.append(str(int(minv+i*step))+'-'+str(int(minv+(i+1)*step)))
    plt.xticks(np.linspace(0,len(acnt)-1,len(acnt)),lx)
    
    
    plt.xlabel('分数区间',fontsize=15)
    plt.ylabel('各分数区间客户占比',fontsize=15)
        
    plt.title('psi:{:.6f}(训练集和测试集对比)'.format(psi),fontsize=20)
    
    ax = plt.axes()
    from matplotlib.ticker import FormatStrFormatter
    
    #设置刻度字体和大小
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    labels = ax.get_xticklabels() + ax.get_yticklabels() 
    [label.set_fontname('Times New Roman') for label in labels]
    
        #旋转x轴标题
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
        
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f%%'))
    plt.show()
    return psi,minv,maxv,step
```

```python
#计算lift、绘制Lift曲线
#将count_list倒序排列（按分数由高到低排序）
count_list = count_list.sort_index(ascending=False)
badrate_random=map(lambda x:x*100/pd.Series(count_list).sum(),count_list) 
badrate_random=list(badrate_random)

#将每区间的坏人数按倒序排列（按分数由高到低排序）
bad_num = table_score_test_label.label.sort_index(ascending=False)
badrate_model=map(lambda x:x*100/(table_score_test_label.label).sum(),bad_num) 
badrate_model=list(badrate_model)

cum_badrate_random=np.cumsum(badrate_random)
cum_badrate_model=np.cumsum(badrate_model)

lift=cum_badrate_model/cum_badrate_random

lift_x=range(len(count_list))

lift_table=pd.DataFrame({'obs':count_list,
                         'bad':bad_num,
                         'bad(%)captured by model':badrate_model,
                         'bad(%)captured randomly':badrate_random,
                         'cum bad(%)captured by model':cum_badrate_model,
                         'cum bad(%)captured randomly':cum_badrate_random,
                         'lift':lift})

lift_table
```

```python
def categ_feature_univariate_analy(feature,pred_prob,true_label,xlabel_name,y_label_name1,y_label_name2,title_name):
    feature_prob_label = pd.concat([feature.reset_index(drop=True),pd.Series(pred_prob),true_label.reset_index(drop=True)], axis=1)
    feature_prob_label.columns = [xlabel_name,y_label_name1,y_label_name2]

    # 变量和评分的关系
    table_feature_prob_label = pd.pivot_table(feature_prob_label, values=[y_label_name1,y_label_name2], index=[xlabel_name],aggfunc=np.mean)

    fig=plt.figure(figsize=(6,4))
    ax1=fig.add_subplot(111)
    ax1.plot(table_feature_prob_label.index,table_feature_prob_label[y_label_name1],'-',color='b',label=y_label_name1)

    ax2 = ax1.twinx() # this is the important function  
    ax2.plot(table_feature_prob_label.index,table_feature_prob_label[y_label_name2],'-',color='r',label=y_label_name2)

    ax1.set_xlabel(xlabel_name)
    ax1.set_ylabel(y_label_name1)
    ax2.set_ylabel(y_label_name2)

    fig.legend(loc=1,bbox_to_anchor=(0.3,1),bbox_transform=ax1.transAxes)

    plt.title(title_name)

    #设置数据分类名称
    plt.xticks(table_feature_prob_label.index,list(pd.value_counts(feature).sort_index().index))
    #旋转x轴标题
    #plt.xticks(rotation=90)
    for tick in ax1.get_xticklabels():
        tick.set_rotation(90)

    plt.show()
```

```python
def num_feature_univariate_analy(feature,pred_prob,true_label,xlabel_name,y_label_name1,y_label_name2,title_name):
    group_names = pd.cut(feature,10).values.categories.left
    feature_bin = pd.cut(feature,10,labels=group_names)
    count_list = DataFrame(feature_bin).groupby(xlabel_name).size()

    feature_bin = DataFrame(np.array(feature_bin))
    #test_y = DataFrame(test_y.reset_index(drop=True))
    feature_prob_label = pd.concat([feature_bin,pd.Series(pred_prob),true_label.reset_index(drop=True)], axis=1)
    feature_prob_label.columns = [xlabel_name,y_label_name1,y_label_name2]
    table_feature_prob_label = pd.pivot_table(feature_prob_label, values=[y_label_name1,y_label_name2], index=[xlabel_name],aggfunc=np.mean)

    fig=plt.figure(figsize=(6,4))
    ax1=fig.add_subplot(111)
    ax1.plot(table_feature_prob_label.index,table_feature_prob_label[y_label_name1],'-',color='b',label=y_label_name1)
    
    
    ax2 = ax1.twinx() # this is the important function  
    ax2.plot(table_feature_prob_label.index,table_feature_prob_label[y_label_name2],'-',color='r',label=y_label_name2)

    ax1.set_xlabel(xlabel_name)
    ax1.set_ylabel(y_label_name1)
    ax2.set_ylabel(y_label_name2)

    fig.legend(loc=1,bbox_to_anchor=(0.3,1),bbox_transform=ax1.transAxes)

    plt.title(title_name)

    #设置数据分类名称
    #plt.xticks(table_feature_prob_label.index,list(pd.value_counts(feature).sort_index().index))
    #旋转x轴标题
    #plt.xticks(rotation=90)
    for tick in ax1.get_xticklabels():
        tick.set_rotation(90)

    plt.show()
```
