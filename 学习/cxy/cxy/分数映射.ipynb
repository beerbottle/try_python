```python
import pandas as pd
import numpy as np
%matplotlib inline 
def score_ys(data,p,middle_score,outlier_pct,score_):
    score_cutoff=np.mean(data[p]) #平均预测概率，假设平均预测概率=实际坏账率
    score_cutoff=np.log(score_cutoff/(1-score_cutoff)) #平均预测概率对应的logodds值
    middle_score=middle_score #可比分数
    outlier_pct = outlier_pct #极端值临界点
    data['odd']=np.log(data[p]/(1-data[p])) #logodds值
    max_odds =np.percentile(data['odd'],100-outlier_pct)#99分位点对应的logodds值
    min_odds=np.percentile(data['odd'],outlier_pct)#1分位点对应的logodds值
    max_p=np.percentile(data[p],100-outlier_pct)#99分位点对应的P值
    min_p=np.percentile(data[p],outlier_pct)#1分位点对应的p值
    ##第一段公式
    b1 =(900-850)/(0-min_p)
    l1 = b1*data[p] + 900
    ##第二段公式
    b2 = (850-middle_score)/(min_odds -score_cutoff )
    a2 = middle_score - b2 * score_cutoff
    l2 = a2 + b2 * data['odd']
    ##第三段公式
    b3 = (middle_score-350)/(score_cutoff -max_odds )
    a3 = middle_score - b3 * score_cutoff
    l3 = a3 + b3 * data['odd']
    ##第四段公式
    b4 = (350-300)/(max_p-1)
    a4 = 300-b4
    l4 = a4+b4*data[p]
    data[score_]=np.where(data['odd']<=score_cutoff,l2,l3)
    data[score_]=np.where(data['odd']> max_odds,l4,data[score_])
    data[score_]=np.where(data['odd']<min_odds ,l1,data[score_])
    data[score_].hist(bins=30)#分数的直方图
    
    x1 = score[(score.odd <= min_odds)]
    x2 = score[(score.odd > min_odds) & (score.odd<=score_cutoff)]
    x3 =  score[(score.odd > score_cutoff) & (score.odd<=max_odds)]
    x4 = score[(score.odd > max_odds)]
    
    fig = plt.figure(figsize=(12,8))
    #ax1 = fig.add_subplot(111)
    plt.plot(x1.phat,x1.score_,'or--',label = u'l1')
    plt.plot(x2.phat,x2.score_,'bs',label = u'l2')
    plt.plot(x3.phat,x3.score_,'gs',label = u'l3')
    plt.plot(x4.phat,x4.score_,'oy-',label = u'l4')
    plt.legend()
```

```python
# %load 分值计算(1).py
"""
Created on Mon Nov 25 13:38:01 2019

@author: dell-3450-i3
"""

#裴博士要求：
#1.分数可比。比如每个评分卡的600分，对应的坏账率大概都是一个值
#2.分数必须在一个区间内，比如300，900
#解决思路：
#1.针对第一个要求，对每个模型计算不同分数段上，对应的坏账率的值。比如(0.8934,0.9145]对应的坏账率为0.21，那么假定
#score_cutoff=(0.8934+0.9145)/2=0.90395,对应的逾期为0.2，分数映射结果为middle_score=600
#2.针对第二个要求，把分数按照middle_score分为2部分，分别按照PDO的方式映射到[300,middle_score],[middle_score,900]。
#因为PDO的映射涉及到log转换，0，1是没法计算。所以对极端值，（极端值的定义为按照百分比，大于或者小于outlier_pct就认定为极端值）
########################################################################
import pandas as pd
import os
import numpy as np
import sys
sys.path.append(r'G:\RiskModelTools')
from publictools.model.score_bin import score_bin
from publictools.model.calc_cutoffvalue   import calc_cutoffvalue
from publictools.model.value_fit   import value_fit
########################################################################
data_dir=r'D:\文档\交付文档\线上api模型-2019-11-06（同盾下线）没有badratemap变量\output_files'
#score=pd.read_pickle(os.path.join(data_dir,'test_ds.pkl'))
#flag=pd.concat(
#        [
#                pd.read_pickle(os.path.join(data_dir,'training.pkl')),
#                pd.read_pickle(os.path.join(data_dir,'testing1.pkl')),
#         ]
#        )[['flag8_minage2_overduedays30','intoid']]
#score_test=pd.merge(score,flag,on='intoid',how='inner')
#score_test.to_pickle(os.path.join(data_dir,'score_test.pkl'))
score_test=pd.read_pickle(os.path.join(data_dir,'score_test.pkl'))
########################################################################

def cal_score(
        indata,
        middle_score,
        yname,
        outlier_pct =1,#==0.01
        bad_rate=0.1
        ):
    score_=indata.copy()
    score_['odd']=np.log(score_['p']/(1-score_['p']))
    lvar=pd.unique([round(s,3) for s in calc_cutoffvalue(var= score_['p'].tolist(),percent_num=10*[10])])
    score_=value_fit(x= pd.Series(score_['p']) ,predefined_bins=lvar) .join(score_)
    cutoff_values = score_bin(indata=score_, scorevar='_labels_',yname=yname, vrange=lvar) 
    cutoff_values =cutoff_values.sort_values(by=['bad_rate']).reset_index(drop=True)
    cutoff_values=cutoff_values.loc[cutoff_values['bad_rate']>bad_rate,'bin_range'].tolist()[0]
    score_cutoff_p=round(eval(cutoff_values.split(',')[0].split("(")[1]+"+"+cutoff_values.split(',')[1].split("]")[0])/2,4)
    ########################################################################
    max_score,min_score   =np.percentile(score_['p'],[100-outlier_pct,outlier_pct])
    ########################################################################
    more_b=(850-middle_score)/(np.log(max_score/(1-max_score))-np.log(score_cutoff_p/(1-score_cutoff_p)) )
    more_a=middle_score-np.log(score_cutoff_p/(1-score_cutoff_p))* more_b
    
    less_b=(middle_score-350)/(np.log(score_cutoff_p/(1-score_cutoff_p))-np.log(min_score/(1-min_score)))
    less_a=middle_score-np.log(score_cutoff_p/(1-score_cutoff_p))* less_b
    
    max_b =(900-850)/(1-max_score)
    max_a =900-max_b
    min_a =300
    min_b =(350-300)/min_score
    ########################################################################
    #总的
    print(max_score)
    print(min_score)
    print(score_cutoff_p)
    bin1= (score_['p']>=score_cutoff_p)&(score_['p']<=max_score)
    bin2= (score_['p']< score_cutoff_p)&(score_['p']>=min_score)
    bin3= (score_['p']>max_score)
    bin4= (score_['p']<min_score)
    
    score_.loc[bin1,'score']=np.log(score_.loc[bin1,'p']/(1-score_.loc[bin1,'p']))*more_b+more_a
    score_.loc[bin2,'score']=np.log(score_.loc[bin2,'p']/(1-score_.loc[bin2,'p']))*less_b+less_a
    score_.loc[bin3,'score']=score_.loc[bin3,'p']*max_b+max_a
    score_.loc[bin4,'score']=score_.loc[bin4,'p']*min_b+min_a
    
    return score_[['score']]


score_test1=cal_score(
        indata=score_test,
        middle_score=600,
        yname='flag8_minage2_overduedays30')    
score_test1['score'].hist()
 

```
